```python
from qce24_qre_challenge.sparse_matrix import BandedCirculantMatrix
from qce24_qre_challenge.quantum_oracle import BandedCirculantOracle
from qce24_qre_challenge.circuit import BlockEncoding
```

```python
matrix_size = 8
```

Creating a banded circulant matrix with random entries:

```python
matrix = BandedCirculantMatrix(size=matrix_size)
```

The following code creates the quantum circuits for block-encoding the banded circulant matrix. For details of this circuit implementation, see https://doi.org/10.1137/22M1484298.

```python
banded_circulant_oracle = BandedCirculantOracle(matrix)
```

Drawing the quantum circuit for the oracle which locates the non-zero matrix entries.

Note that Qiskit's qubit ordering is reversed compared to the aforementioned paper due to its big-endian convention.

```python
location_oracle_circuit = banded_circulant_oracle.location_oracle()
location_oracle_circuit.draw('mpl')
```

Drawing the quantum circuit for the oracle which encodes the matrix entries:

```python
value_oracle_circuit = banded_circulant_oracle.value_oracle()
value_oracle_circuit.draw('mpl')
```

Drawing the full block-encoding circuit:

```python
block_encoding = BlockEncoding(banded_circulant_oracle)

block_encoding_circuit = block_encoding.create_circuit()
block_encoding_circuit.draw('mpl')
```
