```python
from azure.quantum.qiskit import AzureQuantumProvider
from qiskit import QuantumCircuit
```

```python
provider = AzureQuantumProvider(workspace)
backend = provider.get_backend('microsoft.estimator')
```

```python
from pathlib import Path

path = Path.cwd().parents[1].joinpath('qasm_circuits/block_encoding_circuits')
circuit_dict = {}
for file_path in list(path.iterdir()):
    with open(file_path, "r") as file:
        circuit_dict[file_path.name] = file.read()
```

```python
qiskit_circuit = QuantumCircuit.from_qasm_str(circuit_dict['block_encoding_4'])

job = backend.run(qiskit_circuit)
result = job.result()
result
```
