```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x_matrix_sizes = [4, 8, 16, 32, 64, 128, 256]
y_logical_qubits = [20, 23, 25, 28, 30, 33, 35]
y_physical_qubits = [333000, 334350, 284450, 304180, 273660, 556950, 591710]

x_remove_anomalies = [4, 8, 128, 256]
y_physical_remove_anomalies = [333000, 334350, 556950, 591710]
```

```python
a_logical, b_logical = np.polyfit(np.log(x_matrix_sizes), y_logical_qubits, 1)
print(f"Logical qubits fit: {round(a_logical, 3)}*log(x)+{round(b_logical, 3)}")

a_physical, b_physical = np.polyfit(np.log(x_remove_anomalies), y_physical_remove_anomalies, 1)
print(f"Physical qubits fit: {round(a_physical, 3)}*log(x)+{round(b_physical, 3)}")
```

```python
print(a_logical, b_logical)
print(a_physical, b_physical)
```

```python
x = np.linspace(4, 256, 100)
y_logical_qubits_fitted = [a_logical*np.log(t) + b_logical for t in x]
y_physical_qubits_fitted = [a_physical*np.log(t) + b_physical for t in x]
```

```python
plt.plot(x_matrix_sizes, y_logical_qubits)
plt.plot(x, y_logical_qubits_fitted)
plt.show()
```

```python
plt.plot(x_matrix_sizes, y_physical_qubits)
plt.plot(x, y_physical_qubits_fitted)
plt.show()
```
