```python
import pyqsp
from pyqsp.poly import StringPolynomial
from pyqsp.angle_sequence import QuantumSignalProcessingPhases, Polynomial
ang_seq = QuantumSignalProcessingPhases([0, 0, 0, 0, 0.5], signal_operator="Wx")
print(ang_seq)
poly = StringPolynomial("0.01/x", 1)
pyqsp.response.PlotQSPResponse(ang_seq, target=poly, signal_operator="Wx")
```

```python
pg = pyqsp.poly.PolyOneOverX()
pcoefs = pg.generate()
poly = Polynomial(pcoefs)

ang_seq = QuantumSignalProcessingPhases(
    poly, signal_operator="Wx", method="laurent", measurement="x"
)
pyqsp.response.PlotQSPResponse(ang_seq, target=poly, signal_operator="Wx", measurement="x")
```
